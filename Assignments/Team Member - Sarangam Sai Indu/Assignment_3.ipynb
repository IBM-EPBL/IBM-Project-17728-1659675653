### Deal with outliers

```python
data = dataset.iloc[:,:-1]
data.boxplot( rot = 90, figsize=(25,10))
```

```python
outlier_variables = ['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight','Viscera weight', 'Shell weight']
for i in outlier_variables:
        q1=dataset[i].quantile(0.25)
        q3=dataset[i].quantile(0.75)
        iqr=q3-q1
        upper=q3+1.5*iqr
        lower=q1-1.5*iqr
        dataset[i]=np.where(dataset[i] >upper, upper, dataset[i])
        dataset[i]=np.where(dataset[i] <lower, lower, dataset[i])
```

```python
data = dataset.iloc[:,:-1]
data.boxplot( rot = 90, figsize=(25,10))
```

## BUILD REGRESSION MODEL

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=6)
xpoly = poly.fit_transform(X)
```

### fit the model

```python
from sklearn.linear_model import LinearRegression
polyreg = LinearRegression()
polyreg.fit(xpoly, y)
```

### prediction

```python
predn = polyreg.predict(poly.fit_transform(X))
```

### performance metrics
